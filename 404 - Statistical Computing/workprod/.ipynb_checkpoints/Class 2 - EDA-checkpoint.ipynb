### Observations that need to be explored further
* There is a small subset of stores who are missing the second half of 2014. It would be good to look into these stores more to see if they have any commonalities or if this is just an issue of random missingness (data error). [Table0](#Rows_per_Store)

### Trends that jump out
* DayOfWeek: There is a very strong DayOfWeek seasonality where Sunday has lower number of Sales. Mondays look to have the highest Sales per store. [Table1](#DayOfWeek_StoreMean) This is in part, due to the fact that many stores are closed on Sundays. [Table2](#DayOfWeek_StoreOpen) The Monday may be higher due to the store was closed on Sunday and people need food after the closure.

### Potential modeling approaches
* The original goal of this kaggle competition is to predict sales for 6 weeks at the day-store level (1k+ stores). I am only familiar with predicting on a couple of timeseries and have never approached anything at this scale. I would like to do some more research into Deep Learning to see if there are any appropriate methods there. Alternatively, I think you can set this problem using a tree-based method like GBM/XGBoost/RF.
* One thing I will need to keep in mind is to not randomly split but to randomly pick a point and then do a pre/post as the train/test split.

### Summary of data exploration
* Promos: Promos only occurred on Monday - Friday. Even after controlling for the DayOfWeek, we see that the sales per store are much higher on Promos day. [Figure1](#DoW_Promos)
* School Holidays do not seem to make a significant difference in the Sales. [Figure2](#DoW_SchoolHoliday)
* Looking at the distribution of Daily Sales, there is a large portion of Days where the value is 0 (probably due to the stores being closed). The non-0 days do not look to be too badly skewed. [Figure3](#DailySales_dist)
* Looking at the trend for all stores across time, nothing sticks out particularly to me. [Figure4](#Sales_by_Day)
* Looking at the sales per store, most of the stores fall in between having an average Sales volume of 2,500 to 10,000 per day. This distribution shows right skew. [Figure5](#Sales_per_Store_mean)

### Description of Data
There are 3 different csvs that contain data
1. store.csv : store level data
2. train.csv : store-day level data

Store.csv has some information on the store type as well as some basic information about competitors. I plotted the distribution of the distance of the nearest competitor and found that the distance was right-skew. The 25th and 75th percentile of competitor distance was 717-6800 meters away which equates to 0.4-4.2 miles. This table acts more as a dimension table.

Train.csv is a fact table and has information at the store-day level. It has information about whether the store was open that day, it was running a promotion, and whether it was a State or School Holiday.
