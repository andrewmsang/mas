```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 100
```

```python
df = pd.read_csv('https://projects.fivethirtyeight.com/polls-page/president_primary_polls.csv')
df.loc[:,'created_at'] = pd.to_datetime(df.loc[:,'created_at'])
# filter down to iowa
iowa = df.loc[(df['state']=='Iowa')&
              (df['cycle']==2020)&
              (df['party']=='DEM'),:]
```

```python
iowa.loc[:,'days_until_primary'] = (pd.to_datetime('2020-02-03') - pd.to_datetime(iowa.loc[:,'created_at'])).dt.days + 1
```

```python
iowa.head()
```

```python
# Which poll and question do we take?
polls = (iowa.copy()
             .loc[:,['question_id','poll_id','pollster_id','created_at']]
             .drop_duplicates())
polls.loc[:,'rnk'] = (polls.groupby(['pollster_id'])['created_at']
                           .transform(lambda x: x.rank(ascending=False)))

first_poll = polls.loc[polls['rnk']==1,:]
first_poll.head()
```

```python
# first preference results
# https://docs.google.com/spreadsheets/d/1mNtJ94lUrKwwX6-q2b_YQvg4EOQ92BsnKiCyLrgrBTo/edit?usp=sharing

fv = pd.read_csv('../data/first_vote.csv')
fv = fv.loc[:,['candidate','first_rd_votes']]
fv.loc[:,'first_rd_share'] = fv['first_rd_votes'] / fv['first_rd_votes'].sum()
fv
```

```python
# filter down to 1 question per pollster
iowa_df = iowa.merge(first_poll, how='inner', on=['question_id','poll_id','pollster_id'])
iowa_df.head()
```

```python
# generate features to get mapped candidate
candidates = fv['candidate'].unique()
iowa_df.loc[:,'mapped_candidate'] = iowa_df['answer'].map(lambda s: s if s in candidates else 'Other')
```

```python
results = iowa_df.merge(fv, left_on=['mapped_candidate'], right_on=['candidate'])
results.head(1)
```

```python
results.head()
```

```python
# exploratory analysis
def rmse(pred, act):
    return np.sqrt(np.mean((pred - act)**2))
    
q_res = results.groupby('question_id').apply(lambda row: rmse(row['pct']/100,row['first_rd_share'])).reset_index()
q_res.rename(columns={0:'rmse'}, inplace=True)

dim = results.loc[:,['question_id','sponsors','pollster_rating_id','fte_grade','sample_size',
                     'population','methodology','days_until_primary']].drop_duplicates()
dim.loc[:,'sample_size_bucket'] = pd.qcut(dim.sample_size, 5)
dim.loc[:,'days_until_bucket'] = pd.qcut(dim.days_until_primary, 5)
dim = dim.merge(q_res, on=['question_id'])
```

```python
dim.groupby('sponsors')['rmse'].agg(['median','count'])
```

```python
dim.groupby('fte_grade')['rmse'].agg(['median','count'])
```

```python
dim.groupby('sample_size_bucket')['rmse'].agg(['median','count'])
```

```python
dim.groupby('methodology')['rmse'].agg(['median','count'])
```

```python
dim.groupby('days_until_bucket')['rmse'].agg(['median','count'])
```

```python
# exploratory analysis
def rmse(pred, act):
    return np.sqrt(np.mean((pred - act)**2))
```

```python
import sqlite3

with sqlite3.connect("../db/votes.db") as conn:
    df = pd.read_sql_query("SELECT * from votes", conn)
```

```python
iowa = pd.read_csv('../data/538_iowa_poll.csv')
iowa.head()
```

```python
metrics = [m for m in df.columns if '_votes' in m]
data = df.groupby('round')[metrics].sum().reset_index()
data = data.loc[data['round']==1,:]
```

```python
dat = pd.melt(data, id_vars='round', var_name='candidate')
dat.loc[:,'candidate'] = dat.loc[:,'candidate'].map(lambda s: (s[:-6]).title())
dat.loc[:,'share'] = dat.groupby('round')['value'].transform(lambda x: x/x.sum())
```

```python
res = iowa.merge(dat, how='inner', left_on=['mapped_candidate'], right_on=['candidate'])
fnl = (res.groupby(['display_name','poll_id'])
          .apply(lambda row: rmse(row['share'], row['pct']))
          .reset_index())
fnl.rename(columns={0:'rmse'}, inplace=True)
fnl.sort_values('rmse').head()
```
